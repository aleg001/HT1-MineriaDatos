```python
conteoVariables = baseballData["home_team"].value_counts()
tam = conteoVariables / conteoVariables.sum()
plt.pie(tam, labels=conteoVariables.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
```

```python
conteoVariables = baseballData["away_team"].value_counts()
tam = conteoVariables / conteoVariables.sum()
plt.pie(tam, labels=conteoVariables.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
```

```python
conteoVariables = baseballData["venue"].value_counts()
tam = conteoVariables / conteoVariables.sum()
plt.pie(tam, labels=conteoVariables.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
```

```python
conteoVariables = baseballData["game_type"].value_counts()
tam = conteoVariables / conteoVariables.sum()
plt.pie(tam, labels=conteoVariables.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
```
